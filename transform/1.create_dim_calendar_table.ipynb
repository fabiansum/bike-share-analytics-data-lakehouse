### Create Calendar Table

```
# Create temp calendar view
from pyspark.sql.functions import explode, sequence, to_date

beginDate = '2013-01-01'
endDate = '2022-12-31'

(
  spark.sql(f"select explode(sequence(to_date('{beginDate}','yyyy-MM-dd'), to_date('{endDate}','yyyy-MM-dd'), interval 1 day)) as calendarDate")
    .createOrReplaceTempView('dates')
)
```

```
%sql
-- Create dim_calendar
DROP TABLE IF EXISTS gold.dim_calendar;
CREATE TABLE IF NOT EXISTS gold.dim_calendar
USING DELTA
AS
SELECT
    CalendarDate as date_id,
    date_part('dayofweek', CalendarDate) as day_of_week,
    date_part('day', CalendarDate) as day_of_month,
    month(CalendarDate) as month ,
    quarter(CalendarDate) as quarter,
    year(CalendarDate) as year
FROM dates
ORDER BY CalendarDate
```

```
%sql
SELECT * FROM gold.dim_calendar
```
